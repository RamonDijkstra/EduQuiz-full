```python
import openai
import os
import json
from transformers import GPT2Tokenizer
```

```python
# Change these variables to perform completion for you task

api_key_path = '<PATH_TO_YOUR_API_KEY_TXT_FILE>'
test_path = os.path.abspath(os.getcwd()).split('gpt3_completion_scripts')[0] + 'processed_data/gpt3/completion_6/processed_test.json'
quiz_path = os.path.abspath(os.getcwd()).split('gpt3_completion_scripts')[0] + 'generated_data_gpt3/experiment_6/generated_quiz.json'
FINE_TUNED_FQ_MODEL = '<NAME_FINE_TUNED_GPT3_COMPLETION_1>'
outfile = os.path.abspath(os.getcwd()).split('gpt3_completion_scripts')[0] + 'generated_data_gpt3/experiment_6/generated_questions_2.json'
```

# Generate questions based on context and answer

```python
with open(api_key_path) as f:
    openai.api_key = f.read()
```

```python
# Open test datasets

test_data = []

for line in open(test_path):
    test_data.append((json.loads(line)))
    
# Get questions
quizzes = []
with open(quiz_path) as f:
    generated_quizzes = json.load(f)

for key in generated_quizzes:
    quizzes.append(generated_quizzes[key])
```

```python
GPT2tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
# The index is also printed below. It can occur that the OpenAI server is overload.
# To overcome the issue of losing all information manually change the for i in range(len(test_data)) line
# For example: the server gets overloaded on number 150
# Change the code to for i in range(151, len(test_data)), comment out generated = {} and continue the process
```

```python
generated = {}

for i in range(len(test_data)):
    answer = quizzes[i].split('True answer: ')[1].split("\nFalse answer")[0]
    prompt = test_data[i]['prompt'].split("###\n\n")[0] + "Answer: " + answer + '\n\n###\n\n' 
        
    if len(GPT2tokenizer(prompt)['input_ids']) + 200 < 2048:
        output = openai.Completion.create(
                    model=FINE_TUNED_FQ_MODEL,
                    max_tokens = 200,
                    temperature = 0,
                    frequency_penalty = 0,
                    presence_penalty =  0,
                    stop = '\n###',
                    prompt = prompt)
        print(i, output['choices'][0]['text'])
        generated[i+1] = output['choices'][0]['text']  
```

```python
with open(outfile, 'w') as f:
    json.dump(generated, f)
```

# Generate answers based on context and question

```python
FINE_TUNED_FQ_MODEL = '<NAME_FINE_TUNED_GPT3_COMPLETION_2>'
outfile_2 = os.path.abspath(os.getcwd()).split('gpt3_completion_scripts')[0] + 'generated_data_gpt3/experiment_6/generated_answers_2.json'
```

```python
generated_2 = {}

for i in range(len(test_data)):
    question = quizzes[i].split(' Question: ')[1].split("\nTrue answer")[0]
    prompt = test_data[i]['prompt'].split("###\n\n")[0] + "Question: " + question + '\n\n###\n\n'
    
    if len(GPT2tokenizer(prompt)['input_ids']) + 200 < 2048:
        output = openai.Completion.create(
                    model=FINE_TUNED_FQ_MODEL,
                    max_tokens = 200,
                    temperature = 0,
                    frequency_penalty = 0,
                    presence_penalty =  0,
                    stop = '\n###',
                    prompt = prompt)
        print(i, output['choices'][0]['text'])
        generated_2[i+1] = output['choices'][0]['text']  
```

```python
with open(outfile_2, 'w') as f:
    json.dump(generated_2, f)
```
